```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import chromedriver_autoinstaller
import time
```

```python
#INSTALL BROWSER SUCH CHROME OR EDGE
chromedriver_autoinstaller.install()

#USE THE INSTALLED BROWSER FOR SELENIUM
browser = webdriver.Chrome()
```

```python
data = []

for index in range (1,30):
    browser.get(f'https://www.uniqlo.com/id/id/men/tops/tops-collections?path=%2C%2C22808{index}')
    time.sleep(10)
    soup = BeautifulSoup(browser.page_source, 'lxml')
    rows = soup.find_all('div', class_='info')
    
    for row in rows:
        color_element = row.find('ul', class_='color-tips')
        color = color_element.text.replace("\n", "").strip() if color_element is not None and color_element.text.strip() != "" else None
        gender_size_element = row.find('div', class_='gender-size fr-flbox')
        gender_size = gender_size_element.text.replace("\n", "").strip() if gender_size_element is not None and gender_size_element.text.strip() != "" else None
        description_element = row.find('h2', class_='description description-text fr-no-uppercase')
        description = description_element.text.replace("\n", "").strip() if description_element is not None and description_element.text.strip() != "" else None
        price_element = row.find('span', class_='fr-price-currency')
        price = price_element.text.replace("\n", "").strip() if price_element is not None and price_element.text.strip() != "" else None
        price_flag_element = row.find('div', class_='price-flag')
        price_flag = price_flag_element.text.replace("\n", "").strip() if price_flag_element is not None and price_flag_element.text.strip() != "" else None
        rating_element = row.find('div', class_='rating')
        rating = rating_element.text.replace("\n", "").strip() if rating_element is not None and rating_element.text.strip() != "" else None
        
        dt = {
            'color': color,
            'gender_size': gender_size,
            'description': description,
            'price': price,
            'price_flag': price_flag,
            'rating': rating
        }
        
        data.append(dt)
```

```python
df = pd.DataFrame(data)
```
